Mapa discreto logístico, la expresión del sistema dinámico es,

$$ x_{n+1} = r x_{n}\left(1-x_{n}\right)$$

Para $r_{onset} = 3.56995$ se tiene comportamiento caótico

```python
import numpy as np
import matplotlib.pyplot as plt
import binascii
import random
n = 8*5 # numero realizacions probadas
r = 4.
```

```python
xo = random.uniform(0, 1) # condicion inicial
iteraciones = [xo]

# iteracion
for i in range(n):
    iteraciones.append(r*iteraciones[-1]*(1-iteraciones[-1]))
```

Visualización trayectoria anterior

```python
vis, ax = plt.subplots(1)

ax.plot(iteraciones)

plt.show()
```

Vamos interpretar la trayectoria como "bits". Si $x_{n} >= y$ , $bit_{n} = 1$, en caso contrario $bit_{n} = 0$.

```python
y = 0.5
cadenaBits = ''
for bit in iteraciones:
    if (bit >= y):
        cadenaBits += '1'
    else:
        cadenaBits += '0'

# Finalmente devolvemos los bits a cadena de texto
cadena = int(cadenaBits, 2)
binascii.unhexlify('%x' % cadena)
```

Voy a emular una trayectoria de atractor con informacion codificada. Por ejemplo la cadena de carácteres "hello" en ascii. 

$$ bin('hello') = '0110100001100101011011000110110001101111'$$

Entonces, en mi emulacion, si hay un bit con valor cero, pues el mapa caótico debe tener un valor que sea menor al valor límite $y$ que se tenga. En caso contrario, debe ser superior. Puedo generar un numero aleatorio entre ambos rangos para "simularlo". 

```python
hola_ascii_bin = '0110100001100101011011000110110001101111'

y = 0.5
trayectoria = []
for digitoBinario in hola_ascii_bin:
    if int(digitoBinario) == 1:
        sim = random.uniform(y,1)
    else:
        sim = random.uniform(0,y)
        
    trayectoria.append(sim)
```

Visualizamos simulacion

```python
sim_vis, sim_ax = plt.subplots()

sim_ax.plot(trayectoria)

plt.show()
```

Vamos a comparar simulacion con resultado real

```python
comp_vis, comp_ax = plt.subplots()

comp_ax.plot(trayectoria)
comp_ax.plot(iteraciones)

plt.show()
```

En la siguiente celda voy a tratar de sincronizar mi mapa logístico con los trayectorias simuladas que representen la cadena de datos que quiero codificar en el sistema dinámico. Para ello hay que introducir una componenente de sincronización. Voy a plantear el siguiente sistema,

$$ x_{n+1} = x_{n}r\left(1-x_{n}\right)  $$

Si el sistema logistico predice $bit = 0$ y simulación $bit = 1$

$$ x_{n+1}^{cor} = x_{n+1}\left(1- \alpha \right) $$

En caso contrario 

$$ x_{n+1}^{cor} = x_{n+1}\left(1+ \alpha  \right) $$

Si ambos bits son iguales, no hacemos nada. La idea no es que que ambas trayectorias coincidan sino que codifiquen misma cadena de bits.

Asi planteamos un paso logistico y luego aplicamos una corrección. Esta "sincronización" la planteamos indefinidamente hasta que se alcance la sincronización. Si eso ocurriera, solo debemos de "ir hacia atras" para qeudarnos con nuestra condicion inicial. Se supone que esta sincronizado cuando el termino de correccion es cero...


```python
# Asignamos a la fuerza de corrección un valor
alfa = .9 # 0.21 es el mejor resultado con 21.643 de 40
r= 4.
numBitsCadena = n
xo = random.uniform(0,1)
sync = False
y = 0.5
trayectoria = []
sims = []
# La sincronizacion debe ser un while
contadorVueltas = 0
contadorSyncs = 0
maxValorSyncs = 0
while sync == False:
    contadorVueltas += 1
    # Recorremos la información codificada en binario las veces que haga falta
    contadorSync = 0
    trayectoria = []
    sims = []
    trayectoria.append(xo)
    for binChar in hola_ascii_bin:
        correccion = 0.
        digitoBinario = int(binChar)
        # Debemos asignar un valor de mapa logistico a este bit. Y es el criterio para considerar bit 1/0
        if digitoBinario == 1:
            sim = random.uniform(y,1)
        else:
            sim = random.uniform(0,y)
        sims.append(sim)
        # Iteramos el mapa logístico
        x1 = r*xo*(1.-xo)
        
        # El "bit" obtenido es el siguiente
        if x1 < y:
            x_bit = 0
        else:
            x_bit = 1
        
        # Calculamos correccion
        if x_bit == digitoBinario:
            # Nada, esta perfecto
            contadorSync += 1
            correccion = 0.
        elif x_bit == 1 and digitoBinario ==0:
            correccion = -alfa*x1 # Hay que "bajar" el valor x1
        else:
            correccion = alfa*x1 # Hay que "subir" el valor x1
            
        x1 += correccion            
        xo = x1
        
        if contadorSync == numBitsCadena:
            sync=True
        else:
            sync=False
        trayectoria.append(x1)
        
    contadorSyncs += contadorSync
    if contadorSync > maxValorSyncs:
        maxValorSyncs = contadorSync
        trayectoriaMax = trayectoria
        simsMax = sims
    print("Tenemos sincronizados :", contadorSync," bits de un total de ",numBitsCadena)
    if (contadorVueltas > 10000):
        contadorSyncs = contadorSyncs/10000.
        break
print("Promedio de sincronizaciones en la sincronizacion :",contadorSyncs)
print("Valor máximo de sincronizaciones es:", maxValorSyncs)
plt.plot(trayectoriaMax)
plt.plot(simsMax)
```
